```python
# Гиперпараметры
input_dim = len(data.columns) - 1
latent_dim = 16
```

```python
# Архитектура VAE
class VAE(keras.Model):
    def __init__(self, input_dim, latent_dim):
        super(VAE, self).__init__()
        self.encoder = keras.Sequential([
            layers.Dense(64, activation='relu'),
            layers.Dense(32, activation='relu'),
            layers.Dense(latent_dim, activation='linear')
        ])
        self.decoder = keras.Sequential([
            layers.Dense(32, activation='relu'),
            layers.Dense(64, activation='relu'),
            layers.Dense(input_dim, activation='sigmoid')
        ])

    def call(self, inputs):
        z = self.encoder(inputs)
        return self.decoder(z)
```

```python
# Обучение модели
vae = VAE(input_dim, latent_dim)
vae.compile(optimizer='adam', loss='mse')
x_train = data.drop(columns=['track_id']).values
vae.fit(x_train, x_train, epochs=10, batch_size=32)
```

```python
# Генерация рекомендаций
encoded_tracks = vae.encoder.predict(x_train)
recommended_tracks = vae.decoder.predict(encoded_tracks)

# Декодирование треков обратно
recommended_track_ids = encoder.inverse_transform(np.argmax(recommended_tracks, axis=1))
print("Рекомендованные треки:", recommended_track_ids[:10])
```
